### Fix Missing Ages

```python
data = [df,df_test]
```

## Check For Relavancy

### Fix Missing Values & Add Attributes to both training/test datasets

```python
for dataset in data:
    # Replace rows with NaN in "Age" with mean age of respective Sex
    dataset["Age"] = dataset[["Age"]].fillna(dataset.groupby(["Sex"]).transform(np.mean))
    
    # Replace rows with NaN in "Fare" with mean fare
    #dataset["Fare"] = dataset[["Fare"]].mean()
    # Combine SibSp and Parch columns into 1 column
    dataset['party-size'] = dataset['SibSp'] + dataset['Parch']
    
    # Setup dummies for Sex column
    dataset = pd.concat([dataset,pd.get_dummies(dataset[["Sex"]])], axis=1)
data[0].head()
```

```python
missing_data = data[1].isnull()
#missing_data.head(5)
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts()) 
    print("")
```

### Drop Irrelevant Columns

```python
df.drop(["Cabin","Name", "Sex","PassengerId", "Survived","SibSp","Parch","Ticket","Cabin","Embarked", "Fare"], axis=1, inplace=True)
df_test.drop(["Cabin","Name", "Sex","PassengerId","SibSp","Parch","Ticket","Cabin","Embarked","Fare"], axis=1, inplace=True)
df.head()
```

```python
missing_data = df_test.isnull()
#missing_data.head(5)
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts()) 
    print("")
```

## Setup & Train Model

```python
decisionTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
```

```python
X_train = df
y_train = df_survived
X_train.head()
```

```python
decisionTree.fit(X_train, y_train)
```
